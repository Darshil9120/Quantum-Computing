```python
circuit1 = QuantumCircuit(4,2)
circuit1.x((0,1))
circuit1.barrier()
circuit1.cx(0,2)
circuit1.cx(1,2)
circuit1.ccx(0,1,3)
circuit1.barrier()
circuit1.measure([2],[0])
circuit1.measure([3],[1])
%matplotlib inline
circuit1.draw(output='mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
result = execute(circuit1, backend = backend, shots=1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
